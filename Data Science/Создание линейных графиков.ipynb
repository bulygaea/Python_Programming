```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(start=-3., stop=3., num=1000)
y1 = np.exp(x)
y2 = 10 + 0.4*x - 0.3*x**2 + 0.1*x**3
```

```python
fig = plt.figure(figsize=(7,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y1, linestyle='--')
axes.plot(x[::30], y2[::30], 'gD')
```

# Гистограммы распределения признаков

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# import warnings; warnings.simplefilter('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('tips.xls')
```

```python
sns.set()
sns.distplot(df.total_bill)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind='hex')
```

```python
sns.pairplot(df)
```

# Работа с категориальными признаками

```python
sns.countplot(x = 'day', data = df)
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df, estimator = min)
```

```python
sns.boxplot(x = 'day', y = 'tip', data = df, hue = 'smoker')
```

```python
sns.pairplot(df, hue = 'smoker')
```

# Графики табличного типа

```python
correlation = df.corr()
```

```python
sns.heatmap(correlation, annot = True, cmap = 'coolwarm')
```

# Основные типы графиков в Plotly

```python
import pandas as pd
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
%matplotlib inline
```

```python
init_notebook_mode(connected = True)
cf.go_offline()
```

```python
df = pd.read_csv('tips.xls')
```

```python
df.plot()
```

```python
df[['total_bill', 'tip']].iplot(kind = 'spread')
```

```python
df.iplot(kind = 'bubble', x = 'total_bill', y = 'size', size = 'tip')
```
